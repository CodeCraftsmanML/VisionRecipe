## Representation Of Images 

### 1. Why Study Images ?

Image helps machines interpret, analyze and make dicision based on visual data 

### 2. Basic of Pixels

Image are represented by grid of pixels. 

Each pixel store colour information. 

The number of pixels define the resolution of image. 

### 3. Basics of RGB 

RGB is a colour model used to represent and display images through combination of red, green and blue light. 

### 4. Basics of Channels 

Red, Green, and Blue are the channels for an image. 

### 5. What is Noise ? 

Random values generated from range (0,255) in a given size 

### 6. Image representation in pyton : Pillow 

A python library for opening, manipulating, and saving various image file formats

Easy-to-use function for opening image, resizing, cropping, addint texts and more

```python
import  numpy as np 
from PIL import Image, ImageDraw, ImageFont
import openpyxl
from openpyxl.styles import PatternFill
import openpyxl.workbook

image_path = "S:\\Machine Learning\\Siddharthan Playlist\\GEN AI for Computer Vision\\data\\GettyImages-653001154-e1691965000531.webp"
image = Image.open(image_path)
image = image.convert('RGB')

# Resizing the image
image = image.resize((64, 128))

# Pixel data
pixels = list(image.getdata())
width, height = image.size

wb = openpyxl.Workbook()
ws = wb.active

for y in range(height):
    for x in range(width):
        r, g, b = pixels[y * width + x]
        cell = ws.cell(row=y + 1, column=x + 1)
        cell.value = f"{r}, {g}, {b}"
        # Corrected format string for hexadecimal color
        fill = PatternFill(start_color=f'{r:02X}{g:02X}{b:02X}', end_color=f'{r:02X}{g:02X}{b:02X}', fill_type='solid')
        cell.fill = fill

excel_path = 'S:\\Machine Learning\\Siddharthan Playlist\\GEN AI for Computer Vision\\data\\shiba.xlsx'
wb.save(excel_path)
```

```python
noise_width, noise_height = 256, 256 

noise = np.random.randint(0,256, (height, width, 3),dtype = np.uint8) 

noise_image = Image.fromarray(noise, 'RGB') 

noise_image_path = 'S:\\Machine Learning\\Siddharthan Playlist\\GEN AI for Computer Vision\\data\\noise_image.png'
noise_image.save(noise_image_path) 
```

```python
import matplotlib.pyplot as plt 
# Basics of Pillow library 

# 1. opening a image 
image_path = 'S:\\Machine Learning\\Siddharthan Playlist\\GEN AI for Computer Vision\\data\\computer_vision.jpg' 
image = Image.open(image_path) 
print(image.size) 
# image.show() 
plt.imshow(image) 
plt.axis(False) 
```

```python
# 2. Resizing an image 
new_size = (200,200) 
resized_image = image.resize(new_size) 
# resized_image.show() 
plt.imshow(resized_image) 
plt.axis(False) 
```

```python
# 3. Cropping an image 

crop_box = [10, 100, 210, 120] 
cropped_image = image.crop(crop_box) 
plt.imshow(cropped_image) 
plt.axis(False) 
```

```python
# 4. adding text to an image 

text_image = image.copy() 
draw = ImageDraw.Draw(text_image) 
text = "Computer Vision"
font = ImageFont.load_default() 
text_position = (100,100) 
draw.text(text_position,text,fill = 'red', font = font) 
plt.imshow(text_image) 
plt.axis(False) 
```

```python
# 5. Rotation of Image 

rotate_image = image.rotate(45) 
plt.imshow(rotate_image) 
plt.axis(False) 
```

```python
# Converting to GrayScale 

gray = image.convert("L") 
plt.imshow(gray,cmap='gray') 
plt.axis(False) 
```
